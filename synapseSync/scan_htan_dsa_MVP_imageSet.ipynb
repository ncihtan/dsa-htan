```python
## This will go to the MVP image set collection created for the DSA/HTAN integration
## This script will demonstrate how to get imaging/item metadata from something that has been indexed on the DSA
### Will also demonstrate how we can POST metadata to a specific image/item as well as how to pull internal imaging
### details
import girder_client
import girder_utils as gu
import secrets as s  ## Put the API Key in here
import pprint
gc = girder_client.GirderClient(apiUrl="https://imaging.htan.dev/girder/api/v1")
gc.authenticate(apiKey=s.dsaApiKey)
```

```python
## All of the MVP images have been copied/created in betaTestImages which is also the MVP Image set

MVP_collection =  gc.get("collection?text='betaTestImages'")[0]
## Let's get the list of folders currently available for the MVP... One onion though I will point out in a minute


MVP_root_folder_list = list(gc.listFolder(MVP_collection['_id'],parentFolderType='collection'))
## This is a girderism, but because I am looking at the ROOT folder, it's parent is actually a collection not a folder


internalMetadataDict = {}

for f in MVP_root_folder_list:
    ## I only care about the folders that start with HTAN, the other one I created manually
    ## as a set of virtual folders to show off some sorting capabilities
    if 'HTAN' in f['name']:
        ## Now let's see who lives in here..
        print(f['name'])
        subFolders = list(gc.listFolder(f['_id']))
        
        for sf in subFolders:
            print("\t",sf['name'])
            ## Let's make this easy for Sheila.. I just want the darn items...
            itemSet = gu.recurseGetItems(gc,sf['_id'])
            for i in itemSet:
                print("\t\t",i['name'])
                try:
                    internalMetadata = gc.get("/item/%s/tiles/internal_metadata" % i['_id'])
                except:
                    internalMetadata = "AintGotAny"
                internalMetadataDict[i['name']] = internalMetadata
```
